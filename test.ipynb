```python
import pandas as pd
df = pd.read_csv("results.csv", header = None, names=["device_token", "event_type", "event_date", "received_date", "controller_date", "c1", "c2", "c3"])
df["total_duration"] = df["controller_date"]-df["event_date"]
df["received_duration"] = df["received_date"]-df["event_date"]
df["controller_duration"] = df["controller_date"]-df["received_date"]
df[["controller_duration", "received_duration", "total_duration"]].plot()
```
